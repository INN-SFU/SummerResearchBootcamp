### 4. Creating a virtual environment

If we can just install packages directly into our python environment why can't we just start analyzing our data?

TLDR. Conflicts between dependencies.

If you are wokring on multiple projects at the same time there is a non-zero chance that they depend on different versions of packages or potentially different versions of python itself. This problem is espescially important in science where we value the reproducibility of results. To avoid these issues we can use virtual environment.

A virtual environment is a self-contained location that enables you to maintain separate and isolated python environments for you python projects.

In practice this means we can install whatever packages we want into our projects and try out difference packages as we develop the code without affecting other working projects. This also means that we can safely delete the virtual environment without affecting any other python projects.

In an ideal world, you would have a "clean" installation of python on your OS and be able to return to that clean slate each time you want to create a new project avoiding any issues with dependencies.

```python
%%bash

virtualenv myenv
source myenv/bin/activate
pip install numpy mne
which python
pip list
deactivate
pip list
```
